Una tienda de jugetes tiene dos productos estrella, un payaso y un juego de mesa, el payaso pesa 150 gr y el juego de mesa pesa 170 gr. hacer una funcion que reciba el numero de payasos y el numero de juegos de mesa por pedido, si el peso total del pedido es mayor a 10 kg debe de mandar error, debe de indicar por cuantos gramos se pasa el envio y deve regresar false al programa principal,

```python
def pedido(payasos, mesa):
    pesop = payasos * 150
    pesom = mesa * 170
    suma = pesop + pesom
    if (suma > 10000):
       return (suma - 10000,False)
    else:
       return (suma,True)

a = int (input("cantidad de payasos: "))
b = int (input("cantidad de juegos de mesa: "))

res,status = pedido (a,b)

if (status):
  print("El peso total es de",res,"gr",status)
else:
    print("Peso exedido por",res,"gr",status)
```
